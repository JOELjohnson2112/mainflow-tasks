```python
#1 What factors are most strongly correlated with a diabetes diagnosis?
data_numeric = data.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
correlation = data_numeric.corr()['Diagnosis'].sort_values(ascending=False)
top_corr = correlation[1:11]  # Top 10 correlated factors excluding 'Diagnosis' itself

sns.barplot(x=top_corr.values, y=top_corr.index, palette='coolwarm')
plt.title('Top 10 Correlations with Diabetes Diagnosis', fontsize=14)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.show()
```

```python
#2 How do BMI and physical activity level influence diabetes diagnosis and associated risk factors?

plt.figure(figsize=(12, 6))
sns.scatterplot(x='BMI', y='FastingBloodSugar', hue='Diagnosis', data=data_numeric, palette='viridis', alpha=0.7)
plt.title('BMI vs. Fasting Blood Sugar by Diagnosis', fontsize=14)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Fasting Blood Sugar', fontsize=12)
plt.legend(title='Diagnosis', labels=['Non-Diabetic', 'Diabetic'])
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='PhysicalActivity', y='HbA1c', hue='Diagnosis', data=data_numeric, palette='coolwarm', alpha=0.7)
plt.title('Physical Activity vs. HbA1c by Diagnosis', fontsize=14)
plt.xlabel('Physical Activity', fontsize=12)
plt.ylabel('HbA1c', fontsize=12)
plt.legend(title='Diagnosis', labels=['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
#3 Do socioeconomic status and education level affect medication adherence and health literacy?

plt.figure(figsize=(12, 6))
sns.boxplot(x='Diagnosis', y='MedicationAdherence', hue='SocioeconomicStatus', data=data_numeric, palette='pastel')
plt.title('Medication Adherence by Diagnosis and Socioeconomic Status', fontsize=14)
plt.xlabel('Diagnosis', fontsize=12)
plt.ylabel('Medication Adherence', fontsize=12)
plt.legend(title='Socioeconomic Status', bbox_to_anchor=(1.05, 1))
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Diagnosis', y='HealthLiteracy', hue='EducationLevel', data=data_numeric, palette='muted')
plt.title('Health Literacy by Diagnosis and Education Level', fontsize=14)
plt.xlabel('Diagnosis', fontsize=12)
plt.ylabel('Health Literacy', fontsize=12)
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
#4 What role do lifestyle factors play in predicting diabetes or related complications?
lifestyle_factors = ['Smoking', 'AlcoholConsumption', 'DietQuality', 'SleepQuality']
sns.pairplot(data_numeric, vars=lifestyle_factors, hue='Diagnosis', palette='coolwarm', diag_kind='kde')
plt.suptitle('Pairwise Relationships of Lifestyle Factors', y=1.02, fontsize=16)
plt.show()
```
